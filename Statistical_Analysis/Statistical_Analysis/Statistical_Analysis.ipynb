***b) Is there any evidence that the players overall rating is greater than 80?***

```python
#H0: mu<=80
#H1: mu>80
alpha = 0.05

#Normality Test
t,p=st.shapiro(dfn['Overall_rating'])

if p<alpha:
    print("Reject the Null Hypothesis || Not a normal Distribution")
else:
    print("Fail Reject the Null Hypothesis || Is a normal Distribution")
    
#So,we will use the non-parametric test   

median=dfn['Overall_rating'].median()

#H0: median<=80
#H1: median>80

x=dfn['Overall_rating']-80.0

statistic,p_value=st.wilcoxon(x,alternative='greater')   

if p_value<alpha:
    print('P_value:',p_value)
    print("Reject the H0 || Median of Overall Rating is greater than 77 ")
else:
    print('P_value:',p_value)
    print("Fail to Reject the H0 || Median of Overall Rating is less than or equal to 77 ")
```

***c) Perform the statistical test to identify whether preferred foot has any impact on potential. Also, perform the graphical analysis.***

```python
#H0: mu_left=mu_right
#H1: mu_left!=mu_right

alpha=0.05

left=dfn[dfn['Preferred Foot']=='Left']['Potential']
Right=dfn[dfn['Preferred Foot']=='Right']['Potential']

#Normality Test
stat,p_value=st.shapiro(dfn['Potential'])

if p_value<alpha:
    print('P_value:',p_value)
    print("Reject H0 || Not a Normal Distribution")
else:
    print('P_value:',p_value)
    print("Fail to Reject H0 || It is a Normal Distribution")
    
#So,we will use the non-parametric test

#H0: median_left = median_right
#H1: median_left != median_right

s,p=st.mannwhitneyu(left,Right)

if p<alpha:
    print('P_value:',p)
    print("Reject the H0 || Preffered Foot has the impact on the Potential")
else:
    print('P_value:',p)
    print("Fail to Reject the H0 || Preffered Foot has no impact on the Potential")
    
    
plt.figure(figsize=(8,5),dpi=100)    
sns.kdeplot(left,shade=True,legend=True,Label='Left',linewidth=3)
sns.kdeplot(Right,shade=True,legend=True,Label='Right',linewidth=3)
plt.axvline(left.median(),color='Yellow',ymax=0.72,linestyle='--')
plt.axvline(Right.median(),color='Red',ymax=0.72,linestyle='--',linewidth=2)
plt.title("Potential Vs Preffered Foot",fontsize=15)
plt.xlabel('Potential',size=10)
plt.ylabel('Density',size=10)
plt.legend()
sns.set_style('dark')
plt.show()

print("""->In Graph we can see that the median of Left and Right is same.As the Left(Red) and Right(Yellow) coincide with each other
 Hence, Through Graphical Analysis we conclude that the Preffered Foot has no impact on the Potential.""")
```
